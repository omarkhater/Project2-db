```python
# Let's assume the order of the B+ tree is 4 for these tests.
bpt = BPlusTree(4)

# Inserting multiple elements and visualizing after each insert
elements_to_insert = [(29, 'a'), (5, 'b'), (7, 'c'), (17, 'd'), (19, 'e'), (31, 'f'), (2, 'g'), (23, 'h'), (11, 'i'), (3, 'j')]
for key, value in elements_to_insert:
    bpt.insert(key, value)
    print(f"After inserting {key}:")
    bpt.visualize()
```

```python
# Search for existing and non-existing keys
search_keys = [15, 30, 40]  # 15 and 30 exist, 40 does not
for key in search_keys:
    result = bpt.search(key)
    if result:
        print(f"Key {key} found with value {result}.")
    else:
        print(f"Key {key} not found.")
```
